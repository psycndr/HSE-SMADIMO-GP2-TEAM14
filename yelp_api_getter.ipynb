```python
data["id"].unique().shape
```

```python
data = data.drop_duplicates("id").reset_index()
data
```

```python
data = data[["id", "name", "is_closed", "review_count", "categories", "rating", "transactions", "price"]]
```

```python
transactions_ohe = pd.get_dummies(data["transactions"].explode(), "transactions").groupby(level=0).sum()
```

```python
data = pd.concat([data, transactions_ohe], axis=1)
data
```

```python
data["categories"] = data["categories"].apply(lambda x: [y["title"] for y in x])
```

```python
categories_ohe = pd.get_dummies(data["categories"].explode(), "categories").groupby(level=0).sum()
data = pd.concat([data, categories_ohe], axis=1)
data
```

```python
price_ohe = pd.get_dummies(data["price"])
data = pd.concat([data, price_ohe], axis=1)
```
